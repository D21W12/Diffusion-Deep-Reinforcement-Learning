```python
import os

import ale_py
import gymnasium as gym
import numpy as np
import torch

from project.agents import DDQNAgent
from project.environments import BaseWrapper
from project.environments.loops import TrainingLoop
from project.util.seed import set_seed
```

```python
set_seed(2112)
```

```python
device = "cuda:9"
```

```python
PATH = os.path.join("..", "checkpoints", "dqn", "Breakout_copy.pt")
MEMORY_PATH = os.path.join("..", "checkpoints", "memory", "Breakout_test.pt")
```

```python
env = BaseWrapper.create_environment("ALE/Breakout-v5")
```

```python
agent = DDQNAgent(
        train=True,
        replay_size=10_000,
        n_actions=env.action_space.n,
        replay_start_size=1_000_000,
        device=device,
        final_exploration_frame=1
    )
```

```python
agent = agent.load(PATH)
agent._steps = 0
agent._epsilon = 0.1
```

```python
loop = TrainingLoop(
    env=env,
    agent=agent,
    wandb=False,
)
```

```python
loop.run(10_000)
```

```python
agent.save_memory(MEMORY_PATH)
```
