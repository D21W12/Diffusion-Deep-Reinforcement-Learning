```python
import os
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
import pandas as pd
```

```python
plt.style.use(['science','no-latex', 'nature'])
```

```python
PATH_BREAKOUT = os.path.join("..", "..", "..", "results", "training", "breakout.csv")
PATH_SPACE_INVADERS = os.path.join("..", "..", "..", "results", "training", "space_invaders.csv")
```

```python
breakout = pd.read_csv(PATH_BREAKOUT)
space_invaders = pd.read_csv(PATH_SPACE_INVADERS)
```

```python
breakout.head()
```

```python
space_invaders.head()
```

```python
breakout = breakout.rename(columns={"dry-star-13 - cumulative_reward": "reward"})["reward"]
space_invaders = space_invaders.rename(columns={"brisk-rain-14 - cumulative_reward": "reward"})["reward"]
```

```python
fig, axis = plt.subplots(1, 2, figsize=(10, 3))

xs_brkt = np.arange(breakout.size)
xs_spcin = np.arange(space_invaders.size)

axis[0].plot(xs_brkt, breakout)
axis[0].set_xlabel("episodes")
axis[0].set_ylabel("total reward")
axis[0].set_title("Average Reward on Breakout")

axis[1].plot(xs_spcin, space_invaders)
axis[1].set_xlabel("episodes")
axis[1].set_ylabel("total reward")
axis[1].set_title("Average Reward on SpaceInvaders")
```
