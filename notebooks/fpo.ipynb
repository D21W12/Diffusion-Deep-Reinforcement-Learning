```python
import torch
```

```python
class FloatCounter:

    def __init__(self, i, fpo = 0):
        self._i = i
        self._fpo = fpo

    def __add__(self, other):
        return FloatCounter(self._i + other, self._fpo + 1)
    
    def __mul__(self, other):
        return FloatCounter(self._i * other, self._fpo + 1)
    
    def __div__(self, other):
        return FloatCounter(self._i / other, self._fpo + 1)
```

```python
y = torch.arange(0, 5)
x = torch.tensor([FloatCounter(0), FloatCounter(0), FloatCounter(0), FloatCounter(0), FloatCounter(0),], dtype=FloatCounter)
x * y
```
