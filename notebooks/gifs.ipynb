```python
import os
import ale_py
import gymnasium as gym
from project.environments.loops import TestingLoop
from project.environments import NoiseWrapper, BaseWrapper
from project.agents import DiffusionAgent, DDQNAgent
from project.models import EDMMauMau
```

```python
GAME = "SpaceInvaders"
SIGMA = 0.25
CHECKPOINT_DQN = os.path.join("..", "checkpoints", "dqn", f"{GAME}.pt")
CHECKPOINT_DIFF = os.path.join("..", "checkpoints", "diff", f"{GAME}.pt")
```

```python
device = "cuda:0"
```

```python
gym.register_envs(ale_py)
env = BaseWrapper.create_environment(f"ALE/{GAME}-v5")
env = NoiseWrapper(env, sigma=SIGMA, save_history=True)
```

```python
model = EDMMauMau.from_checkpoint(CHECKPOINT_DIFF, device, sigma_noise=SIGMA).to(device)
agent = DDQNAgent(
    train=False,
    n_actions=env.action_space.n,
).load(CHECKPOINT_DQN).to(device)
```

```python
loop = TestingLoop(
    env=env,
    agent=agent
)
```

```python
loop.run(1200)
```

```python
env.save_history("../results/animation_diff", duration=35)
```
